## Word Frequency and TF IDF
### Step 1: Preprocessing

```python
import re
import nltk
from nltk.corpus import stopwords
doc_text = " "

def preprocess_docs(text):
    text = str(text).lower()
    text = re.sub('[^a-zA-z0-9\s]','',str(text))            
    text = text.split()  
    text = [wl.lemmatize(word) for word in text if not word in stopwords.words('english')]
    new_text = ' '.join(text)
    return new_text
corpus = [text,text2]
text1 = preprocess_docs(text)
text2 = preprocess_docs(text2)
```

### Step 2: Calculating frequency of words in both docs

```python
first= text1
second= text2

first = first.split(" ")
second= second.split(" ")

total= set(first).union(set(second))

wordDictA = dict.fromkeys(total, 0) 
wordDictB = dict.fromkeys(total, 0)
for word in first:
    wordDictA[word]+=1
    
for word in second:
    wordDictB[word]+=1
    
pd.DataFrame([wordDictA, wordDictB])
```

### Step 3: Calculating Term Frequency

```python
def computeTF(wordDict, bow):
    tfDict = {}
    bowCount = len(bow)
    for word, count in wordDict.items():
        tfDict[word] = count/float(bowCount)
    return tfDict

tfFirst = computeTF(wordDictA, first)
tfSecond = computeTF(wordDictB, second)

tf_df= pd.DataFrame([tfFirst, tfSecond])
tf_df.head()
```

### Step 4: Calculating IDF

```python
import math
def computeIDF(docList):
    idfDict = {}
    N = len(docList)
    
    idfDict = dict.fromkeys(docList[0].keys(), 0)
    for doc in docList:
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log10(N / float(val))
        
    return idfDict

idfs = computeIDF([wordDictA, wordDictB])
idfs
```

### Step 5: Calculating TF-IDF value

```python
def computeTFIDF(tfBow, idfs):
    tfidf = {}
    for word, val in tfBow.items():
        tfidf[word] = val*idfs[word]
    return tfidf
#running our two sentences through the IDF:
idfFirst = computeTFIDF(tfFirst, idfs)
idfSecond = computeTFIDF(tfSecond, idfs)
#putting it in a dataframe
idf= pd.DataFrame([idfFirst, idfSecond])
```
