# Assignment A4- Data Visualisation-I

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

dataset = sns.load_dataset('titanic')

dataset.head()
```

The dataset contains 891 rows and 15 columns and contains information about the passengers who boarded the unfortunate Titanic ship.

Source: Inbuilt seaborn dataset

### Identifying patterns in the dataset

#### Dist Plots

```python
sns.histplot(dataset['fare'], kde=True);
```

Thus, most of the tickets have been bought between 0-50 dollars. Maximum tickets were sold at a cheaper rate. In order to get a more accuare number within given range, bins parameter can be used as follows: 

```python
sns.displot(dataset['fare'], kde=False, bins=10);
```

#### Point Plot
A point plot represents an estimate of central tendency for a numeric variable by the position of the dot and provides some indication of the uncertainty around that estimate using error bars.

```python
sns.pointplot(x='sex', y='fare', data=dataset);
```

#### Line Plot
The relationship between x and y can be shown for different subsets of the data using the hue, size, and style parameters with the line plot 

```python
sns.lineplot(x='age', y='fare', data=dataset, hue='survived');
```

#### Joint Plot
The jointplot() is used to display the mutual distribution of each column

```python
sns.jointplot(x='age', y='fare', data=dataset, kind='hex');
```

```python
sns.jointplot(x='sex', y='fare', data=dataset);
```

Thus, most of the tickets for both males and females have been bought at a cheaper rate, especially in males among the age group of 20-35. 

#### Bar Plot
The barplot is used to display the mean value for each value in a categorical column, against a numeric column.

```python
sns.barplot(x='sex', y='fare', data=dataset);
```

From the output, it can be clearly see that the average fare for male passengers is around 70 while the average fare for female passengers is around 90.

#### Box Plot
The box plot is used to display the distribution of the categorical data in the form of quartiles. The center of the box shows the median value. 

```python
sns.boxplot(x='class', y='age', data=dataset, hue="survived");
```

The box plot thus shows that in all 3 classes, majority of people who survived had a lower median age.

#### Violin Plot
The violin plot is similar to the box plot, however, the violin plot allows us to display all the components that actually correspond to the data point. 

```python
sns.violinplot(x='sex', y='age', data=dataset);
```

```python
sns.violinplot(x='class', y='age', data=dataset, hue='survived');
```

The area where the violin plot is thicker has a higher number of instances for the age. Thus, from the violin plot, it is clearly evident that the number of passengers with age between 20 and 40 is higher than all the rest of the age brackets in almost all classes.

The number of people who survived from first class all had a much higher age between 40-60.
The plots can also be combined as follows:


```python
sns.violinplot(x='class', y='age', data=dataset, hue='survived', split=True);
```

#### The Swarm Plot
This function is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap.

```python
sns.swarmplot(x='sex',y='age',data=dataset,hue='survived',dodge=True);
```

#### The Strip Plot
The strip plot draws a scatter plot where one of the variables is categorical. It is similar to a scatter plot, but with mixed categorial and numerical variables.

```python
sns.stripplot(x='sex',y='age',data=dataset,hue='survived',dodge=True);
```

```python
sns.stripplot(x='class',y='fare',data=dataset,hue='survived',dodge=True, jitter=True);
```

Thus, the fares for first class were obviouly much higher. It is interesting to note that only people who paid excessively high fairs in the first class were guaranteed survival.

#### Matrix Plots
Matrix plots are the type of plots that show data in the form of rows and columns. Heat maps are the prime examples of matrix plots.

Heat map showing correlation between all attributes is given as:

```python
corr = dataset.corr()
sns.heatmap(corr, annot=True, cmap='autumn');
```

```python
grids = sns.PairGrid(dataset)
grids.map(plt.scatter);
```

#### Facet Grids
The facet grids are used to plot two or more than two categorical features against two or more than two numeric features.

```python
grid = sns.FacetGrid(data=dataset, col='alive', row='sex')
grid.map(sns.histplot, 'age');
```

This plot shows that the maximum count of people who survived were women in lower age groups. This is consistent with the fact that the principle of "Women and Children First" was followed while loading the lifeboats.

```python
grid = sns.FacetGrid(data=dataset, col='class', row='alive')
grid.map(plt.scatter, 'age','fare');
```

This plot shows that the maximum count of people who who did nt survive paid lower fares and belonged to lower classes of the ship.
