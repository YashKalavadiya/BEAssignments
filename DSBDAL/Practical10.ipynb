```python
import pandas as pd
import docx2txt
import spacy
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from math import log10
```

```python
text = docx2txt.process("testdoc.docx")
```

```python
text2 = docx2txt.process("testdoc2.docx")
```

```python
nlp = spacy.load("en_core_web_md")
```

```python
data1 = nlp(text)
data2 = nlp(text2)
```

```python
len(data1)
```

```python
len(data2)
```

```python
for word in data1:
    print(f"{word.text} ---> {word.pos_} ---> {word.lemma_}")
```

```python
words_without_stop_words_data1 = []
for word in data1:
    if word.text not in stopwords.words():
        words_without_stop_words_data1.append(word.text)
```

```python
len(words_without_stop_words_data1)
```

```python
words_without_stop_words_data2 = [word.text for word in data2 if word.text not in stopwords.words()]
```

```python
len(words_without_stop_words_data2)
```

```python
ps = PorterStemmer()
stemmes_data1 = []
for word in words_without_stop_words_data1:
    stemmes_data1.append(ps.stem(word))
```

```python
stemmes_data2 = []
for word in words_without_stop_words_data2:
    stemmes_data2.append(ps.stem(word))
```

```python
wl = WordNetLemmatizer()
lemmatized_data1 = []
for word in words_without_stop_words_data1:
    lemmatized_data1.append(wl.lemmatize(word))
```

```python
lemmatized_data2 = []
for word in words_without_stop_words_data2:
    lemmatized_data2.append(wl.lemmatize(word))
```

```python
lemmatized_data1
```

```python
tf = []
corpus = [lemmatized_data1, lemmatized_data2]
for data in corpus:
    count = {}
    for word in data:
        if word in count:
            count[word] += 1
        else:
            count[word] = 1
    tf.append(count)
```

```python
for i in range(len(corpus)):
    length = len(corpus[i])
    for word in tf[i].keys():
        tf[i][word] = tf[i][word] / length
```

```python
tf
```

```python
unique_words = set(lemmatized_data1).union(set(lemmatized_data2))
idf = {}
for word in unique_words:
    idf[word] = 0
    for data in corpus:
        if word in data:
            idf[word] += 1
```

```python
len_of_corpus = len(corpus)
for word in idf.keys():
    idf[word] = log10(len_of_corpus / idf[word])
```

```python
tf_idf = []
for doc in tf:
    dict_sent = {}
    for word in doc.keys():
        dict_sent[word] = doc[word] * idf[word]
    tf_idf.append(dict_sent)
```
